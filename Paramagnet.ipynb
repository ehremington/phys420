```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python
def factorial(x):
    return sp.factorial(x, exact=True)

def multiplicityParamagnet(N,up):
    return factorial(N)//factorial(up)//factorial(N-up)

def logArray(array):
    import math as math
    return [math.log(x) for x in array]
```

```python
df = pd.DataFrame({'up':range(100+1)})
```

```python
df['down'] = 100-df['up']
```

```python
df['multiplicity'] = [multiplicityParamagnet(100,up) for up in df['up']]
```

```python
df['entropy'] = logArray(df['multiplicity'])
```

```python
df['energy'] = 100-2*df['up']
```

```python
df.sort_values('up', ascending=False, inplace=True)
```

```python
df['d ent'] = np.gradient(df['entropy'])
```

```python
df['d eng'] = np.gradient(df['energy'])
```

```python
df['temp'] = df['d eng']/df['d ent']
```

```python
#import seaborn as sns
#sns.set()

fig0, (ax0, ax1) = plt.subplots(2,figsize=(8,10))

ax0.plot(df['energy'],df['entropy'], label='entropy')
ax1.plot(df['energy'],df['temp'], label='temp')

ax0.set_ylabel('S/$k_b$')
ax0.set_xlabel('U/$\mu$B')
ax1.set_ylabel('T $k_b$/$\mu$B')
ax0.set_xlabel('U/$\mu$B')
ax0.legend()
ax1.legend()

ax0.grid(True)
ax1.grid(True)
```

```python
df['heat capacity'] = np.gradient(df['energy'])/np.gradient(df['temp'])/100
```

```python
fig2, ax2 = plt.subplots()

ax2.plot(np.abs(df['temp']), df['heat capacity'], label='heat capacity')
```
