```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def p(v,t):
    return 8*t/(3*v-1)-3/(v)**2
```

```python
v = np.linspace(0.5, 4, 100)

fig0 = plt.figure()
ax0 = fig0.add_subplot(111)

ax0.plot(v, p(v,0.8), label=r'$0.8T_c$')
ax0.plot(v, p(v,0.9), label=r'$0.9T_c$')
ax0.plot(v, p(v,1.0), label=r'$T_c$')
ax0.plot(v, p(v,1.1), label=r'$1.1T_c$')
ax0.plot(v, p(v,1.2), label=r'$1.2T_c$')

ax0.legend()

ax0.set_ylabel(r'$P/P_c$')
ax0.set_xlabel(r'$V/V_c$')
```

```python
df = pd.DataFrame({'v/vc':v,
                   '0.8Tc p/pc':p(v,0.8),
                   '0.9Tc p/pc':p(v,0.9),
                   '1.0Tc p/pc':p(v,1.0),
                   '1.1Tc p/pc':p(v,1.1),
                   '1.2Tc p/pc':p(v,1.2),
                  })
```

```python
df.plot(x='v/vc', y='0.9Tc p/pc')
```

```python
df['0.9Tc G'] = -0.9*np.log(3*df['v/vc']-1)+.9/(3*df['v/vc']-1)-9/4/df['v/vc']
```
